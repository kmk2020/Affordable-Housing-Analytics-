```python
# Import the dependencies.
import pandas as pd

import matplotlib.pyplot as plt

import pathlib as Path
```

```python
# Files to load
bottom_tier = "bottom_tier.csv"
top_tier = "top_tier.csv"
```

```python
# Load CSVs
# Load Zillow Home Value Index (ZHVI) CSVS
#higher value homes value index

# top_tier
bottom_tier_df = pd.read_csv(bottom_tier)
bottom_tier_df.head(10)
```

```python
# top_tier
top_tier_df = pd.read_csv(top_tier)
top_tier_df.head(10)
```

```python
# MERGE THE TWO DATAFRAMES

# import module
import pandas as pd
  
# assign dataframes
data1 = bottom_tier_df
  
data2 = top_tier_df
  
# display dataframes
print('Dataframes:')
display(data1)
display(data2)
  
# merge two data frames
print('After merging:')
data_concat_df=pd.concat([data1, data2], axis=0)
```

```python
# check the firts 10 records

data_concat_df.head(10)
```

```python
# drop column StateName data_concat_df (duplicate of RegionName) and Region Type
# data_concat_df.drop(['RegionType', 'StateName'], axis=1, inplace=True)
# data_concat_df.head(10)

data_concat_df.dtypes
```

```python
# Check for uniques
data_concat_df.RegionName.unique()
```

```python
#check first 10 rows
data_concat_df.head(10)
```

```python
#data shape
data_concat_df.shape
```

```python
# melt the dataframe
melted_df = pd.melt(data_concat_df, id_vars=['RegionID', 'RegionType','RegionName', 'StateName','SizeRank'], var_name='Date', value_name='ZHVI')
```

```python
melted_df.head(10)
```

```python
# convert the 'Date' column to datetime format
melted_df['Date'] = pd.to_datetime(melted_df['Date'], format='%Y-%m')
```

```python
melted_df.head(10)
```

```python
# confirm the shape of our transformed dataframe
melted_df.shape
```

```python
# Drop the StateName Column(duplictated RegionName column)
zillow_df = melted_df.drop(['StateName'], axis=1) 
```

```python
# Check for nulls
zillow_df.isnull()
```

```python
# publish dataframes into CSVs
zillow_df.to_csv('master_affordable_housing.csv')
```
