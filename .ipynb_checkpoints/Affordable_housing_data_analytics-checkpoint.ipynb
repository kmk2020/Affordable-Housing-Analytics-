```python
# Import the dependencies.
import pandas as pd

import matplotlib.pyplot as plt

import pathlib as Path
```

```python
# Files to load
bottom_tier = "bottom_tier.csv"
top_tier = "top_tier.csv"
```

```python
# Load CSVs
# Load Zillow Home Value Index (ZHVI) CSVS
#higher value homes value index

# top_tier
bottom_tier_df = pd.read_csv(bottom_tier)
bottom_tier_df.head(10)
```

```python
# top_tier
top_tier_df = pd.read_csv(top_tier)
top_tier_df.head(10)
```

```python
# MERGE THE TWO DATAFRAMES

# import module
import pandas as pd
  
# assign dataframes
data1 = bottom_tier_df
  
data2 = top_tier_df
  
# display dataframes
print('Dataframes:')
display(data1)
display(data2)
  
# merge two data frames
print('After merging:')
data_concat_df=pd.concat([data1, data2], axis=0)
```

```python
# melt the dataframe
melted_df = pd.melt(data_concat_df, id_vars=['RegionID', 'RegionType','RegionName', 'StateName','SizeRank'], var_name='Date', value_name='ZHVI')
```

```python
# convert the 'Date' column to datetime format
melted_df['Date'] = pd.to_datetime(melted_df['Date'], format='%Y-%m')
```

```python
# Drop the StateName Column(duplictated RegionName column)
zillow_df = melted_df.drop(['StateName'], axis=1) 
```

ARIMA ML MODEL 
 
    Use ARIMA model to forecast ZHVI 

```python
# Import Pandas Libraries for ARIMA

# fit an ARIMA model and plot residual errors
from pandas import datetime
from pandas import read_csv
from pandas import DataFrame
from statsmodels.tsa.arima.model import ARIMA
from matplotlib import pyplot
```

```python
# Plot the graph
zillow_df.plot()
```

```python
import numpy as np, pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})
```

```python
df1 = pd.DataFrame(zillow_df, columns = ['Date','ZHVI'])
```

```python
df1.head()
```

```python
# Original Series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(df1.ZHVI); axes[0, 0].set_title('Original Series')
plot_acf(df1.ZHVI, ax=axes[0, 1])
```

```python
# 1st Differencing
axes[1, 0].plot(df1.ZHVI.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df1.ZHVI.diff().dropna(), ax=axes[1, 1])
```

```python
# 2nd Differencing
axes[2, 0].plot(df1.ZHVI.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df1.ZHVI.diff().diff().dropna(), ax=axes[2, 1])

#plt.show()
```
